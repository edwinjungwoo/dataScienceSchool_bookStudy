# 데이터프레임 합성

```python
import numpy as np
import pandas as pd

df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007], 
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns = ['고객번호', '이름'])

df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how = 'outer')
```

```python
df1 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎길이': [1.4, 1.3, 1.5, 1.3]},
    columns=['품종', '꽃잎길이'])
df1
```

```python
df2 = pd.DataFrame({
    '품종': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]},
    columns=['품종', '꽃잎너비'])
df2
```

```python
pd.merge(df1, df2)
```

```python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
```

```python
pd.merge(df1, df2, on = "고객명")
```

```python
df1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적2': [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on = '이름', right_on = '성명')
```

```python
df1 = pd.DataFrame({
    '도시': ['서울', '서울', '서울', '부산', '부산'],
    '연도': [2000, 2005, 2010, 2000, 2005],
    '인구': [9853972, 9762546, 9631482, 3655437, 3512547]})
df1
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape((6, 2)),
    index=[['부산', '부산', '서울', '서울', '서울', '서울'],
           [2000, 2005, 2000, 2005, 2010, 2015]],
    columns=['데이터1', '데이터2'])
df2
```

```python
pd.merge(df1, df2, left_on = ['도시', '연도'], right_index = True)
```

```python
df1 = pd.DataFrame(
    [[1., 2.], [3., 4.], [5., 6.]],
    index=['a', 'c', 'e'],
    columns=['서울', '부산'])
df1
```

```python
df2 = pd.DataFrame(
    [[7., 8.], [9., 10.], [11., 12.], [13, 14]],
    index=['b', 'c', 'd', 'e'],
    columns=['대구', '광주'])
df2
```

```python
pd.merge(df1, df2, how = 'outer', left_index=True, right_index=True)
```

```python
df1.join(df2, how = 'outer')
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['데이터1', '데이터2'])
df1
```

```python
data1 = []

for i in range(6):
    data1.append(np.random.randint(1,101, size = 2))
data1
```

```python
data2 = []

for i in range(6):
    data2.append(np.random.randint(1,101, size = 2))
data2
```

```python
df1 = pd.DataFrame(data1, columns = ["매출", "비용"], 
                   index = ['1월', '2월', '3월', '4월', '5월', '6월'])
df1["이익"] = df1["매출"] - df1["비용"]
df1
```

```python
df2 = pd.DataFrame(data2, columns = ["매출", "비용"], 
                   index = ['7월', '8월', '9월', '10월', '11월', '12월'])
df2["이익"] = df2["매출"] - df2["비용"]
df2
```

```python
df3 = pd.concat([df1, df2])
```

```python
df3
```

```python
df3.loc['총실적'] = df3.sum()
df3
```
