```python
columns = ["반", "번호", "국어", "영어", "수학"]

df_score3 = pd.DataFrame([["A", "01", 100, 98, 100], 
                         ["A", "02", 92, 87, 93], 
                         ["A", "03", 88, 100, 95],
                         ["A", "04", 91, 79, 100], 
                         ["A", "05", 88, 99, 88], 
                         ["B", "01", 90, 78, 100], 
                         ["B", "02", 92, 67, 83], 
                         ["B", "03", 90, 100, 67],
                         ["B", "04", 98, 79, 80], 
                         ["B", "05", 88, 79, 81]],
                        columns = columns)

df_score3
```

```python
df_score4 = df_score3.set_index(["반","번호"])
df_score4
```

```python
df_score4["평균"] = round(df_score4.mean(axis=1),2)
df_score4
```

```python
temp = df_score3.set_index(["번호", "반"])
df_score5 = temp.unstack("반").sort_index(level = 0)
df_score5
```

```python
df_score5.loc[("평균", slice(None))] = df_score5.mean()
df_score5
```
