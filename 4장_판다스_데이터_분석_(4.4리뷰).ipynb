```python
import pandas as pd
import numpy as np
```

```python
titanic['category1'] = titanic.apply(lambda x: x.sex if x.age > 20 else "child", axis = 1)
titanic.tail()
```

```python
titanic.age = titanic.age.fillna(titanic.age.mean())
titanic.age = titanic.age.astype(int)
titanic.tail()
```

```python
titanic.age = titanic.age.astype(str)
titanic['category2'] = titanic.sex + titanic.age
titanic.tail()
```

```python
titanic.age = titanic.age.astype(int)

bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
cats = pd.cut(titanic.age, bins, labels = labels)
cats
```

```python
cats.value_counts() / len(cats)
```

```python
titanic["cats"] = cats
```

```python
titanic['category3'] = titanic.apply(lambda x: "미성년자" if x.age < 20 
                                     else x.cats + "남성" if x.sex == "male"
                                     else x.cats + "여성", axis = 1)
titanic.tail()
```
