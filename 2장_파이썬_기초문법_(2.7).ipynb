# 파이썬 for반복문

```python
for i in range(10):
    print("=")
```

```python
for i in range(10):
    print("=" + str(i) + "=")
```

```python
for i in range(10):
    print("*" * i)
```

```python
for i in range(10):
    print("*" * (10-i))
```

```python
for i in range(4):
    for j in range(4):
        print(i + j, end= " ")
    print()
```

```python
for i in range(6):
    n1 = i + 1
    for j in range(6):
        n2 = j + 1
        n = n1 + n2
        if n % 4 == 0:
            print(n1, n2)
```

```python
stock1 = 1024
for i in [0.5, 2]:
    stock2 = stock1 * i
    for j in [0.5, 2]:
        stock3 = stock2 * j
        for k in [0.5, 2]:
            stock4 = stock3 * k
            print(stock4)
```

```python
n = 3
for a in range(1,11):
    for b in range(1,11):
        for c in range(1,11):
            print((a ** n) + (b ** n) == (c ** n))
```

```python
stock1 = 1024
for i in [0.5, 2]:
    for j in [0.5, 2]:
        for k in [0.5, 2]:
            print(stock1 * i * j * k)
```

```python
n = 4

for j in range(n):
    sum = 0
    for i in range(j + 1):
        sum = sum + (i + 1)
    print(sum)
```
