```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
def peak_to_peak(x):
    return x.max() / x.min()

iris.groupby(iris.species).agg(peak_to_peak)
```

```python
iris.groupby(iris.species).mean()
```

```python
iris.groupby(iris.species).describe().T
```

```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3]

iris.groupby(iris.species).apply(top3_petal_length)
```

```python
def top5_petal_width(df):
    return df.sort_values(by = "petal_width", ascending = False)[:5]

iris.groupby(iris.species).apply(top5_petal_width)
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels = ["small", "medium", "large"]).astype(str)

iris["petal_length_class"] = iris.groupby(iris.species).petal_length.transform(q3cut)
iris[["petal_length", "petal_length_class"]].tail(10)
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.tail()
```

```python
labels = ['미성년', '성인', ' 노인']
bins = [1, 20, 60, 100]
titanic['age_class'] = pd.cut(titanic['age'], bins, labels = labels)
titanic.head()
```

```python
titanic.pivot_table('survived', ['sex', 'age_class'], 'class', margins = True)
```

```python
titanic.pivot_table('survived', 'sex', 'class', margins = True)
```

```python
titanic.groupby(["sex", "age_class", "class"])[["survived"]].mean().unstack("class")
```
